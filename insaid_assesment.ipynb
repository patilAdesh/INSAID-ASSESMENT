### Import the required modules and packages

```python
import pandas as pd
import numpy as np
from numpy import array, bool_
```

### Load the required dataset file(s)

```python
df = pd.read_csv('Fraud.csv')
```

```python
df
```

### Drop irrelevant columns

```python
df = df.drop(['step', 'nameOrig', 'nameDest'], axis=1)
```

### Identify the missing values and replace them with 'NaN'

```python
df.replace(' ', np.nan, inplace=True)
```

### Drop rows with missing values

```python
df.dropna(inplace=True)
```

### Identify and remove outliers using IQR Method

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
df = df[~((df < (Q1 - 1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))).any(axis=1)]
```

### Identify and remove multicollinearity by dropping highly correlated features



```python
corr_matrix = df.corr().abs()
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
to_drop = [column for column in upper.columns if any(upper[column] > 0.95)]
df.drop(to_drop, axis=1, inplace=True)
```
